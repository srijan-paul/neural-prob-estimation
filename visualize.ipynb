```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
fig, ax = plt.subplots()


def normal(x: float, sd: float, mean: float) -> float:
    return np.exp(-0.5 * ((x - mean) / sd) ** 2) / (sd * math.sqrt(2 * np.pi))

def generate_normal_dist(range_, sd, mean):
    """ Generate a random distrbution in range [min_x, max_x, step]"""
    [min_x, max_x, n_samples] = range_
    xs = np.linspace(min_x, max_x, n_samples)
    ys = np.array([normal(x, sd, mean) for x in xs])
    return xs, ys

# generate artificial training data
xs, ys = generate_normal_dist([120, 1250, 12500], sd = 120, mean = 430)
ax.plot(xs, ys)
```
